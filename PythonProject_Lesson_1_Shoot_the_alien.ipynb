# Event

Games involves lots of human keyboard presses and mouse clicks. These are events that a program can't predict when they happens. Therefore, games uses a programming technique called **event driven architecture**, that envolves

* Background code monitoring if an event has happened, and 
* User defined function to call when the event did happen. 

The user defined function is called **callback function**

Not just game design uses event driven architecture. GUI design, physical computing also uses it, for example, our Microbit program often involves defining a on_button_pressed() function. That's a callback function for button pressed event. 

Let's look at a frequently used callback function in PyGame Zero. 

````python
def on_mouse_down(pos):
    if alien.collidepoint(pos):
        print("Good Job!")
        place_alien()
    else:
        print("You missed")
````

Here we define a callback function ```on_mouse_down(pos)```. This is a pre-defined callback function name. Whenever a player press down a mouse button, PyGame Zero will go look for this function and run it. Not just the function name is pre-defined, the input parameters name are also pre-defined. So ```on_mouse_down(location)``` won't work. 

Our callback function does a few thing

* ```alien.collidepoint(pos)``` is a method that check if the alien is clicked
* ```place_alien()``` is a function that replace the alien sprite at a new location.
* The ```if...else...``` logic gives different text feedback to the player
