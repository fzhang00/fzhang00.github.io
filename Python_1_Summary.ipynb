# 2. Data Structure

Python has many data structure. We've learn list and dictionary.

## 2.1 List

- List is defined by items enclosed by square bracket, seperated by comma
- List has fixed order
- List items can be mixed type
- List can be modified

```
a_list = ['text1', 'text2', 'text3']
a_num_list = [1,2,3,4,5,65]
# display the third item in a_list
a_list[2]
```

## 2.2 Dictionary

* Defined by key:value pair enclosed by curly bracket
* Each item has a key and value separated by colon
* Key cannot be duplicated. Usually keys are strings. 
* Value can be any data type, ie. string, number, list, function, object, etc.
* Dictionary is changeable, expandable


```
a_dict = {'a':1, 'b': 345, 'c':563458}
b_dict = {'Canada':'Capital', 'China':'Beijing'}
b_dict['Japan'] = 'Tokyo'
a_dict['a']
```

# 7. Function

```
# Define a function
def my_addition(num1, num2, num3):
  addition_result = num1+num2+num3
  return addition_result

# A script that get user input, and use the function to calculate
num=[]
for i in range(3):
  a_num = input("Enter a number:")
  num.append(int(a_num))
my_result = my_addition(num[0], num[1], num[2])
print("The sum of ", num, " is ", my_result)
```

# 9. Module

Python modules gives Python powers to do all kinds of things, from complex text manipulation, time keeping, to web automation and AI

```
import math
print("\n***** Example of math module *****")
print("The biggest common divider for 45 and 63 is:", math.gcd(45,63))
print("E to the power of 64 is:", math.exp(64))

import datetime
print("\n***** Example of datetime module *****")
print("The current time is ", datetime.datetime.today())
```

## 10.2 Open file to write (overwrite)

This will open the file to write. Whatever already in the file will be overwritten

```
# This cell won't run because file doesn't exist on Google server
with open("your_file_name", 'w') as file:
    file.write("Text you want to write to file.")
```

## 10.3 Open File to append

This will add to the end of the file, leaving content already in the file untouched. 

```
with open("your_file_name", 'a') as file:
    file.write("Text you want to append to file.")
```
